### 1. Check if a number is positive, negative, or zero
**Return:** A string such as `'Positive'`, `'Negative'`, or `'Zero'`

```python
def check_number(n: int) -> str:
    if n > 0:
        return f"Positive"
    elif n < 0:
        return f"Negative"
    else:
        return f"Zero"

# check_number(5)
```

```python
# Assertions
assert check_number(5) == 'Positive'
assert check_number(-3) == 'Negative'
assert check_number(0) == 'Zero'
```

### 2. Return a list of numbers from 1 to n (where n can be anything).
Example if n = 5 then
**Return:** `[1, 2, 3, 4, 5]`

```python
def numbers_1_to_n(n) -> list[int]:
    result = []
    for i in range(1, n + 1):
        result.append(i)
    return result

# numbers_1_to_n(5)
```

```python
assert numbers_1_to_n(5) == [1, 2, 3, 4, 5]
```

### 3. Return only even numbers from 1 to n (where n can be anything).
Example if n = 11 then
**Return:** `[2, 4, 6, 8, 10]`

```python
def even_numbers_1_to_n(n) -> list[int]:
    evens = []
    for i in range(1, n + 1):
        if i % 2 == 0:
            evens.append(i)
    return evens

# even_numbers_1_to_n(11)
```

```python
assert even_numbers_1_to_n(11) == [2, 4, 6, 8, 10]
```

### 4. Return the square of a given number
**Example:** For input `4`, return `16`

```python
def square_number(n: int) -> int:
    return n * n

# square_number(5)
```

```python
assert square_number(4) == 16
assert square_number(0) == 0
```

### 5. Return `'Access Granted'` if the password is `'python'`, else `'Access Denied'`
**Example:** Input `'java'` → `'Access Denied'`; Input `'python'` → `'Access Granted'`

```python
def check_password(password: str) -> str:
    if password == 'python':
        return 'Access Granted'
    else:
        return 'Access Denied'

# check_password('python')
```

```python
assert check_password('python') == 'Access Granted'
assert check_password('PYTHON') == 'Access Denied'
assert check_password('java') == 'Access Denied'
```

### 6. Return the larger of two numbers  
**Example:** Input `(4, 9)` → `9`  
**Return:** The larger number.

```python
def larger_number(a: int, b: int) -> int:
    if a > b:
        return a
    else:
        return b

# larger_number(7, 2)
```

```python
# Assertions
assert larger_number(4, 9) == 9
assert larger_number(10, 3) == 10
assert larger_number(5, 5) == 5
```

### 7. Return `'Even'` if a number is even, else `'Odd'`
**Example:** Input `4` → `'Even'`

```python
def even_or_odd(n: int) -> str:
    if n % 2 == 0:
        return 'Even'
    else:
        return 'Odd'

# even_or_odd(10)
```

```python
assert even_or_odd(2) == 'Even'
assert even_or_odd(3) == 'Odd'
```

### 8. Return the sum of numbers from 1 to `n`

**Example:** Input `5` → `15`
because 1+2

```python
def sum_1_to_n(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# sum_1_to_n(7)
```

```python
assert sum_1_to_n(5) == 15
assert sum_1_to_n(10) == 55
```

### 9. Return a list of squares for numbers 1 to n
if n is 5
**Return:** `[1, 4, 9, 16, 25]`

```python
def squares_1_to_n(n) -> list[int]:
    squares = []
    for i in range(1, n+1):
        squares.append(i * i)
    return squares

# squares_1_to_n(5)
```

```python
assert squares_1_to_n(5) == [1, 4, 9, 16, 25]
```

### 10. Count how many times a letter appears in a word  
**Example:** Input `('banana', 'a')`
**Return:** `3`

```python
def count_letter(word: str, letter: str) -> int:
    count = 0
    for ch in word:
        if ch == letter:
            count += 1
    return count

# count_letter('hello', 'l')
```

```python
# Assertions
assert count_letter('banana', 'a') == 3
assert count_letter('apple', 'p') == 2
assert count_letter('python', 'z') == 0
```

### 11. Return the average of three numbers  
**Example:** Input `(4, 8, 10)` → `7.3333...`  
**Return:** The arithmetic mean of the three numbers.

```python
def average_of_three(a: float, b: float, c: float) -> float:
    return (a + b + c) / 3

# average_of_three(1, 2, 3)
```

```python
# Assertions
assert round(average_of_three(4, 8, 10), 2) == 7.33
assert average_of_three(3, 3, 3) == 3
assert average_of_three(10, 0, 5) == 5
```
