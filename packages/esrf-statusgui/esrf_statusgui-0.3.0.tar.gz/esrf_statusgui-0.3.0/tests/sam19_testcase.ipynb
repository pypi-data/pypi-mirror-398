Automate launching post-processing from the GUI interface

Go until the end of usually GUI, you can automate as much as you want every step of the GUI. Do not launch the processing at the final step

```python
import sys
from esrf_pathlib import ESRFPath as Path

from esrf_statusgui.visualization.DCT.pySetupH5 import pySetupH5

launch_gui = pySetupH5()
launch_gui.display()
launch_gui.welcome_tab.widget_exp_select.value = "ma6062"
launch_gui.navigation.tabs.selected_index = 1
launch_gui.dataset_selection_tab.selected_file = "/data/visitor/ma6062/id11/20240711/RAW_DATA/Ti35_1_A_comp/Ti35_1_A_comp_dct5/Ti35_1_A_comp_dct5.h5"
launch_gui.navigation.tabs.selected_index = 2
launch_gui.parameters.acq.nof_phases.value = 1
launch_gui.parameters_tab.widget_energy.value = 43.6
launch_gui.navigation.tabs.selected_index = 3
launch_gui.phases_tab.phase_tab.children[0].composition_dropdown.value = "Ni"
launch_gui.phases_tab.phase_tab.children[0].name_input.value = "Ni"
launch_gui.phases_tab.phase_tab.children[0].material_input.value = "Ni"
launch_gui.phases_tab.phase_tab.children[0].cif_output.children[
    0
].selected_file = "/data/id11/archive/file_xop/In718/Ni.cif"
launch_gui.phases_tab.phase_tab.children[0].reflection_output.children[
    0
].selected_file = "/data/id11/archive/file_xop/In718_ep7_dct3/Ni/Ni.csv"
```

Change the name of the folder for processing

```python
from datetime import datetime

launch_gui.parameters.acq.collection_dir.value = (
    launch_gui.dataset_selection_tab.selected_file.replace(
        "RAW_DATA", "PROCESSED_DATA"
    ).split(".h5")[0]
    + datetime.today().strftime("%Y_%m_%d")
    + ".h5"
)
launch_gui.parameters.acq.name.value = (
    datetime.today().strftime("%Y_%m_%d")
    + launch_gui.dataset_selection_tab.selected_file.replace(
        "RAW_DATA", "PROCESSED_DATA"
    )
    .split(".h5")[0]
    .split("/")[-1]
)
launch_gui.parameters.acq.dir.value = (
    launch_gui.dataset_selection_tab.selected_file.replace(
        "RAW_DATA", "PROCESSED_DATA"
    ).split(
        launch_gui.dataset_selection_tab.selected_file.split("/")[-1].split(".h5")[0]
    )[0]
    + launch_gui.parameters.acq.name.value
)
```

Launch pre-prcoessing from this command at last

```python
launch_gui.setup_pre_processing()
```
