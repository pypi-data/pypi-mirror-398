# Oil Discrimination (Basic)
Load synthetic oil data, run FoodSpec protocol, and preview results.

```python
import pandas as pd
from foodspec.protocol_engine import ProtocolConfig, ProtocolRunner
from pathlib import Path

data_path = Path('examples/data/oil_synthetic.csv')
proto_path = Path('examples/protocols/EdibleOil_Classification_v1.yml') if (Path('examples/protocols/EdibleOil_Classification_v1.yml')).exists() else None
df = pd.read_csv(data_path)
cfg = ProtocolConfig.from_file(proto_path) if proto_path else ProtocolConfig.from_dict({"name": "synthetic_oil", "steps":[{"type":"rq_analysis","params":{"oil_col":"oil_type","matrix_col":"matrix","heating_col":"heating_stage"}}, {"type":"output","params":{"output_dir":"runs_notebook"}}]})
runner = ProtocolRunner(cfg)
res = runner.run([df])
print('Run dir:', res.run_dir)
# Show first lines of report
print(res.report.split('
')[:20])
```
