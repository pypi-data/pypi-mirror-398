# Erythroid Dataset - LSD Training

This notebook demonstrates training the Latent State Dynamics (LSD) model on the Erythroid differentiation dataset.

**Dataset Information:**
- ~9,815 cells
- ~5,000 highly variable genes
- Linear differentiation pathway
- Expected CBDir: ~0.592

## Setup

```python
import os
import scanpy as sc
import numpy as np
import torch

from sclsd import LSD, LSDConfig, set_all_seeds, clear_pyro_state

SEED = 42
clear_pyro_state()
set_all_seeds(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

## Load Data

```python
data_path = "../../data/Erythroid/normlized_before.h5ad"
adata = sc.read(data_path)
print(f"Loaded {adata.n_obs} cells x {adata.n_vars} genes")
```

```python
sc.pl.umap(adata, color='clusters', title='Erythroid - Cell Types')
```

## Configure LSD Model

Key hyperparameters for Erythroid:
- `V_coeff = 5e-3`
- `kl_schedule.af = 3`
- `lr = 2e-3`
- `batch_size = 256`
- `path_len = 16`
- `num_epochs = 150`

```python
cfg = LSDConfig()

cfg.model.V_coeff = 5e-3
cfg.model.z_dim = 10
cfg.model.B_dim = 2

cfg.optimizer.kl_schedule.af = 3
cfg.optimizer.adam.lr = 2e-3

cfg.walks.batch_size = 256
cfg.walks.path_len = 16
cfg.walks.num_walks = 4096
cfg.walks.random_state = SEED
```

## Initialize and Train

```python
lsd = LSD(adata, cfg, device=device)
lsd.set_prior_transition(prior_time_key="palantir_pseudotime")
lsd.prepare_walks()

print(f"Generated {lsd.walks.shape[0]} walks of length {lsd.walks.shape[1]}")
```

```python
model_dir = "./erythroid_model"
os.makedirs(model_dir, exist_ok=True)

lsd.train(
    num_epochs=150,
    save_dir=model_dir,
    save_interval=50,
    random_state=SEED,
)
```

## Extract and Save Results

```python
result = lsd.get_adata()
sc.pl.umap(result, color=['lsd_pseudotime', 'potential', 'entropy'], ncols=3)
```

```python
lsd.save(dir_path=model_dir, file_name="lsd_model_final.pth")
result.write(os.path.join(model_dir, "result_adata.h5ad"))
print(f"Results saved to {model_dir}")
```
